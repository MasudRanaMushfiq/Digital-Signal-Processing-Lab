```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def correlation(x, y):
    N = len(x)
    M = len(y)
    y = y[::-1]
    
    Rxy = np.zeros(N + M - 1)
    
    for n in range(N + M - 1):
        for k in range(N):
            if 0 <= n - k < M:
                Rxy[n] += x[k] * y[n - k]
    return Rxy
```

```python
x = [1, 1, 2, 2]
y = [1, 2, 3, 4]
Rxy = correlation(x, y)
print(Rxy)
```

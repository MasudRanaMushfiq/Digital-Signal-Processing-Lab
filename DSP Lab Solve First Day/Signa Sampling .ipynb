```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 100)
```

```python
def sinosoidal_wave(t):
    return (3*np.sin(2*np.pi*5*t) +
            2*np.cos(2*np.pi*15*t) +
            np.sin(2*np.pi*30*t))
```

```python
y = sinosoidal_wave(t)
```

```python
# plt.stem(t, y)
plt.plot(t, y)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
# plt.show()
```

```python
fs = 50
ts = 1 / fs

ts = np.arange(0, 1, ts)
ys = sinosoidal_wave(ts)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(t, y)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
# plt.stem(ts, ys)
plt.plot(ts, ys)
```

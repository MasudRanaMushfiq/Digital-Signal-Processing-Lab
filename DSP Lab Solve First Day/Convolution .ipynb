```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolution(x, h):
    N = len(x)
    M = len(h)
    y = np.zeros(N + M - 1)
    for n in range(N + M - 1):
        for k in range(N):
            if 0 <= n - k < M:
                y[n] += x[k] * h[n - k]
    return y
```

```python
x = [1, 2, 3, 4]
h = [1, 2, 2, 1]
y = convolution(x, h)
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given x[n] = {1, 2, 3, 4} and find the DFT of the following discrete time signal
```

```python
def DFT(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-1j * 2 * np.pi * k * n / N)
    return X
```

```python
x = [1, 2, 3, 4]
X = DFT(x)
X
```

```python
print(X)
magnitude = np.abs(X)
magnitude
```

```python
phase = np.angle(X)
phase
```

```python
k = np.arange(len(x))

plt.Figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.stem(k, phase)
plt.title("Phase spectrum")
plt.xlabel("K")
plt.ylabel("X[k]")

plt.subplot(1, 2, 2)
plt.stem(k, magnitude)
plt.title("Magnitude spectrum")
plt.xlabel("K")
plt.ylabel("Phase (radian)")
```

```python
# Now perform IDFT operation on the output
```

```python
def IDFT(X):
    N = len(X)
    x = np.zeros(N, dtype=complex)
    for n in range(N):
        for k in range(N):
            x[n] += X[k] * np.exp(1j * 2 * np.pi * n * k / N)
    return x / N
```

```python
reconstruct_x = IDFT(X)
reconstruct_x
```

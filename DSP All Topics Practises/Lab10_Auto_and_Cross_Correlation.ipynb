```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
fs = 200
t = np.arange(0, 1, 1/fs)

a = np.sin(2 * np.pi * 10 * t)             # 10 Hz sine
b = signal.square(2 * np.pi * 10 * t)      # 10 Hz square
c = np.sin(2 * np.pi * 20 * t)     
```

```python
plt.figure()

plt.plot(t, a, label='10 Hz Sine (a)')
plt.plot(t, b, label='10 Hz Square (b)', linestyle='--')
plt.plot(t, c, label='20 Hz Sine (c)', linestyle=':')


plt.title("Signals")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.show()
```

```python
# Auto-correlation of a
auto_a = np.correlate(a, a, mode='full')

# Cross-correlations
cross_ab = np.correlate(a, b, mode='full')
cross_ac = np.correlate(a, c, mode='full')

# Lag axis
lags = np.arange(-len(a)+1, len(a))
```

```python
plt.figure()
plt.plot(lags, auto_a)

plt.title("Auto-correlation of 10 Hz Sine (a)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid()

plt.show()
```

```python
plt.figure()
plt.plot(lags, cross_ab)

plt.title("Cross-correlation between a (sine) and b (square)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid()

plt.show()
```

```python
plt.figure()
plt.plot(lags, cross_ac)

plt.title("Cross-correlation between a (10 Hz) and c (20 Hz)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid()

plt.show()
```

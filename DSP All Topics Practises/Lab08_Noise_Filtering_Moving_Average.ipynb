```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
fs = 100  # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # 1 second duration

# Generate signal: 10 Hz sine wave
signal = np.sin(2 * np.pi * 10 * t)

# Add white noise
noise = np.random.normal(0, 0.5, len(t))
noisy_signal = signal + noise

# Display basic info
print("Number of samples:", len(t))
```

```python
# Define 6-point differencing filter
# Simple differencing: y[n] = x[n] - x[n-1] + x[n-2] - x[n-3] + x[n-4] - x[n-5]

filtered_signal = np.zeros(len(noisy_signal))
print(filtered_signal)

for n in range(len(noisy_signal)):
           filtered_signal[n] = noisy_signal[n] - noisy_signal[n-1] + noisy_signal[n-2] - noisy_signal[n-3] + noisy_signal[n-4] - noisy_signal[n-5]
```

```python
# Plot results
plt.figure(figsize=(12, 6))

plt.plot(t, noisy_signal, label="Noisy Signal", alpha=0.7)
plt.plot(t, filtered_signal, label="Filtered Signal (6-point differencing)", linewidth=2)

plt.title("Noisy vs Filtered Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.show()
```

```python
# 6-point moving average (true smoothing)
kernel = np.ones(6) / 6
smoothed_signal = np.convolve(noisy_signal, kernel, mode='same')

plt.figure(figsize=(12, 6))
plt.plot(t, noisy_signal, label="Noisy Signal", alpha=0.6)
plt.plot(t, smoothed_signal, label="Smoothed Signal (Moving Average)", linewidth=2)

plt.legend()
plt.grid()
plt.title("Smoothing using Moving Average Filter")
plt.show()
```

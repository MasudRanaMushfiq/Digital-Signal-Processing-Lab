```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,1,0])
x2 = np.array([0,1,1,2])
```

```python
y = 3 * x1 - 2 * x2
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(np.arange(len(x1)), x1)
plt.title('Input Signal x1[n]')
plt.subplot(3, 1, 2)
plt.stem(np.arange(len(x2)), x2)
plt.title('Input Signal x2[n]')
plt.subplot(3, 1, 3)
plt.stem(np.arange(len(y)), y)
plt.title('Output Signal y[n] = 3*x1[n] - 2*x2[n]')
plt.tight_layout()
plt.show()
```

```python
Y1 = np.fft.fft(x1)
Y2 = np.fft.fft(x2)
Y = 3 * Y1 - 2 * Y2
y_ifft = np.fft.ifft(y)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.stem(Y1)
plt.title('Signal Y1[n]')
plt.subplot(2, 2, 2)
plt.stem(Y2)
plt.title('Signal Y2[n]')
plt.subplot(2, 2, 3)
plt.stem(Y)
plt.title('Output Signal Y[n] = 3*Y1[n] - 2*Y2[n]')
plt.subplot(2, 2, 4)
plt.stem(y_ifft)
plt.title('Inverse FFT of y_ifft[n]')
plt.tight_layout()
plt.show()
```

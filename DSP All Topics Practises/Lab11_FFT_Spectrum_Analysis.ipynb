```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Parameters
# =========================
fs = 100                     # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)   # Time vector (1 second)

# =========================
# Generate Signals
# =========================
x1 = np.sin(2 * np.pi * 30 * t)   # 30 Hz sine wave
x2 = np.sin(2 * np.pi * 70 * t)   # 70 Hz sine wave

# =========================
# Compute DFT
# =========================
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)

# Magnitude spectra
X1_mag = np.abs(X1)
X2_mag = np.abs(X2)

# Frequency axis
f = np.fft.fftfreq(len(t), d=1/fs)

# =========================
# Plot Signals (Time Domain)
# =========================
plt.figure()

plt.subplot(2,1,1)
plt.plot(t, x1)
plt.title("30 Hz Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(2,1,2)
plt.plot(t, x2)
plt.title("70 Hz Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()

# =========================
# Plot Magnitude Spectra
# =========================
plt.figure()

plt.subplot(2,1,1)
plt.stem(f[:len(f)//2], X1_mag[:len(f)//2])
plt.title("Magnitude Spectrum (30 Hz)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|X(f)|")
plt.grid()

plt.subplot(2,1,2)
plt.stem(f[:len(f)//2], X2_mag[:len(f)//2])
plt.title("Magnitude Spectrum (70 Hz)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|X(f)|")
plt.grid()

plt.tight_layout()
plt.show()

# =========================
# Combined Comparison Plot
# =========================
plt.figure()

plt.stem(f[:len(f)//2], X1_mag[:len(f)//2], label="30 Hz")
plt.stem(f[:len(f)//2], X2_mag[:len(f)//2], label="70 Hz (Aliased)", markerfmt='x')

plt.title("Comparison of Spectra")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|X(f)|")
plt.legend()
plt.grid()

plt.show()
```

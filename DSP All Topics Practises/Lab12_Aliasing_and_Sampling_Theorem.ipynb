```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 100  # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # 1 second time vector

# Define two signals
f1 = 40   # Frequency of first signal (Hz)
f2 = 60   # Frequency of second signal (Hz) -> will alias (since fs/2 = 50)

signal1 = np.sin(2 * np.pi * f1 * t)
signal2 = np.sin(2 * np.pi * f2 * t)

# Combine signals
combined_signal = signal1 + signal2

# FFT
N = len(combined_signal)
fft_vals = np.fft.fft(combined_signal)
fft_vals = np.abs(fft_vals) 

# Frequency axis
freqs = np.fft.fftfreq(N, 1/fs)

# Only take positive frequencies
mask = freqs >= 0
freqs = freqs[mask]
fft_vals = fft_vals[mask]

# Plot time domain
plt.figure(figsize=(12, 4))
plt.plot(t, combined_signal)
plt.title("Combined Signal (Time Domain)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# Plot frequency domain
plt.figure(figsize=(12, 4))
plt.stem(freqs, fft_vals)
plt.title("Magnitude Spectrum (Aliasing Effect)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```

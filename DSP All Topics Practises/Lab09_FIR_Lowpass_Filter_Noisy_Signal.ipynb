```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000
t = np.arange(0, 1, 1/fs)
# t = np.arange(0, 1, 1/1000)
signal = np.sin(2 * np.pi * 10 * t)
# noise1 = np.sin(2 * np.pi * 15 * t)
noise = np.sin(2 * np.pi * 100 * t)

noisy_signal = signal + noise
```

```python
plt.figure()
plt.plot(t, noisy_signal)
plt.title('Noisy Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
fc = 20
N = 51
n = np.arange(N)

# h = (2 * fc / fs) * np.sinc(2 * fc * (n - (N - 1) / 2) / fs)
h = np.zeros(N)
for i in range(N):
    if n[i] == (N - 1) / 2:
        h[i] = 2 * fc / fs
    else:
        h[i] = np.sin(2 * np.pi * fc * (n[i] - (N - 1) / 2) / fs) / (np.pi * (n[i] - (N - 1) / 2))

h *= np.hamming(N)
```

```python
y = np.convolve(noisy_signal, h, mode='same')
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t, noisy_signal)
plt.title('Noisy Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(t, y)
plt.title('Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.tight_layout()
plt.show()
```

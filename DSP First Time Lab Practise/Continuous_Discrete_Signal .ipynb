```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def signal(t):
    return (10 * np.sin(2 * np.pi * 5 * t))
```

```python
t = np.linspace(0, 1, 1000)
```

```python
y = signal(t)
```

```python
plt.plot(t, y)
plt.grid(True)
```

```python
sampling_rate = 5
sampling_time = 1 / sampling_rate

ts = np.arange(0, 1, sampling_time)

ys = signal(ts)
```

```python
plt.stem(ts, ys)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Discrete time signal")
```

```python
plt.Figure(figsize=(16, 10))

plt.subplot(2, 2, 1)
plt.plot(t, y)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Continuous time signal")

plt.subplot(2, 2, 2)
plt.stem(ts, ys)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Discrete time signal")
```
